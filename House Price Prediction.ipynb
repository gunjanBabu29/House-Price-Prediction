## Step 4 : Train-Test Split and feature scaling

```python
# separating target and features for the train_test_split (y=target, x=feature)
y = data[['price']]
x=data.drop(['id', 'price', 'date'],axis=1)
```

```python
#importing library for train_test_split the data
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y, test_size = 0.2,random_state =2)
```

## Step 5 : Model Building

### Ridge Regression
To combat the issue of overfitting in linear regression models, ridge regression is a regularization approach. The size of the coefficients is reduced and overfitting is prevented by adding a penalty term to the cost function of linear regression.

```python
#importing model for Ridge Regression from Sklearn
from sklearn.linear_model import Ridge
```

```python
#model building
ridge = Ridge(alpha=0.001)
ridge.fit(x_train, y_train)
```

```python
#finding total accuracy comes in Ridge Regression
print("Acuuracy of training data=",ridge.score(x_train, y_train)) #accuracy of trained data
print("Acuuracy of testing data=",ridge.score(x_test, y_test)) #accuracy of test data
```

### Lasso Regression
Lasso regression, commonly referred to as L1 regularization, is a method for stopping overfitting in linear regression models by including a penalty term in the cost function

```python
#importing model for Lasso Regression from Sklearn
from sklearn.linear_model import Lasso
```

```python
#model building
lasso = Lasso(alpha=0.01)
lasso.fit(x_train, y_train)
```

```python
#finding total accuracy comes in Lasso Regression
print("Acuuracy of training data=",lasso.score(x_train, y_train)) #accuracy of trained data
print("Acuuracy of testing data=",lasso.score(x_test, y_test)) #accuracy of test datav
```

### Linear Regression 
Linear regression is a type of supervised machine learning algorithm that computes the linear relationship between a dependent variable and one or more independent features. When the number of the independent feature, is 1 then it is known as Univariate Linear regression, and in the case of more than one feature, it is known as multivariate linear regression.

```python
#importing model for LinearRegression from Sklearn
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
#Building model (fit x_train in model)
reg.fit(x_train,y_train)
```

## Step 6 : Accuracy and Evaluation

```python
#finding total accuracy comes in LinearRegression
print("Acuuracy of training data=",reg.score(x_train, y_train)) #accuracy of trained data
print("Acuuracy of testing data=",reg.score(x_test, y_test)) #accuracy of test data
```
